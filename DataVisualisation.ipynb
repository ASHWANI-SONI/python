```python
# line plot 
x = np.arange(10)
y1 = x**2
y2 = 2*x + 3
print(x)
print(y1)
print(y2)
```

```python
plt.plot(x, y1)
plt.show()
```

```python
plt.plot(x, y1, color = "red", label="apple", marker='o')
# plt.show()
plt.plot(x, y2, color='green', label = "kiwi", linestyle = 'dashed')
plt.xlabel('Time')
plt.ylabel('price')
plt.title('Pices of Fruits over Time')
plt.legend()
plt.show()
```

## Scatter Plots

```python
#Adjust the size of of any plot
plt.figure(figsize=(6, 6))
plt.scatter(x, y1, marker='*', label="apple")
plt.scatter(x, y2, marker='o')
plt.legend()
plt.show()
```

```python
plt.bar?
```

```python
plt.bar([0, 1, 2], [10, 20 , 15]) # current year
plt.bar([0, 1, 2], [13, 10 , 12]) # next year
plt.show()
```

```python
x_coordinate = np.array([0, 1, 2])*2
plt.bar(x_coordinate, [10, 20, 15], width=0.5, label='Current year', tick_label=["Gold", "platinum", 'silver']) # Current year
plt.bar(x_coordinate+0.5, [20, 10, 12], width=0.5, label='Next year') # next year
plt.title("metal Price Comparison")
plt.xlabel('Metal')
plt.ylabel('Price')
plt.ylim(0, 40)
# plt.style.use("light_background")
plt.style.use("seaborn-v0_8-paper")
plt.legend()
plt.show()
```
